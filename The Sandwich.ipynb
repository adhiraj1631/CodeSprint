Problem Statement

At Sunshine Elementary, students face a daily challenge during lunch: grabbing their preferred sandwich type from the cafeteria's stack.

Each student prefers either.

0→ circular sandwich

1→ square sandwich

Sandwiches are stacked, and students queue up in order.

The rules are:

The student at the front checks if their preferred sandwich matches the top of the stack

If it matches, they take the sandwich and leave the queue.

If not, they go to the end of the queue.
USC

UPES STUDEN

This continues until no student wants the current top sandwich. Your task is to determine how many students are unable to eat.

Input Format

Line 1: Integer n → number of students and sandwiches.

Line 2: n space-separated integers (0 or 1) → student preferences (queue order).

Line 3: n space-separated integers (0 or 1) → sandwich stack (first element is the top).

Output Format

A single integer: number of students unable to eat.

Example 1

Input

6

101100

010110
Output

0

Explanation

Initial Queue: [1, 0, 1, 1, 0, 0]

Sandwiches (top to bottom): [0, 1, 0, 1, 1, 0]

All students eventually get the sandwich they prefer.

Example 2

Input

4

1100

0000

Output

2

Explanation

Remaining queue: [1, 1]

Remaining sandwiches: [0, 0]
Explanation

Remaining queue: [1, 1]

Remaining sandwiches: [0, 0]

The students left want square sandwiches, but only circular ones remain.

Hence, 2 students go hungry.

Constraints

1 ≤ n ≤ 1000

All input values are either 0 or 1.

```python
from collections import deque

def count_students_unable_to_eat(students, sandwiches):
    queue = deque(students)
    stack = sandwiches
    attempts = 0  # Count of consecutive unmatched students

    while queue and attempts < len(queue):
        if queue[0] == stack[0]:
            queue.popleft()
            stack.pop(0)
            attempts = 0  # reset since a student ate
        else:
            queue.append(queue.popleft())
            attempts += 1

    return len(queue)

# Input reading
n = int(input())
students = list(map(int, input().strip()))
sandwiches = list(map(int, input().strip()))
print(count_students_unable_to_eat(students, sandwiches))
```
