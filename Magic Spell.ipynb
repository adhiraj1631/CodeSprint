In the magical realm of Spellheaven, two wizards, Merlin and Morgana, are preparing for a high-stakes duel.

Each wizard has a set of spells, each with a specific power level.

The duel rules are simple: both wizards randomly select a spell, and the one with the stronger spell wins. If the spells are equally powerful, it's a draw.

Merlin, determined to win, can enhance his spells using mana. Each point of power increase costs 1 mana. Morgana, unaware of this, cannot enhance her spells. Merlin wants to ensure that every possible spell he casts is stronger than any of Morgana's spells, no matter which spells are randomly chosen.
chosen.

Your task is to calculate the minimum mana Merlin needs to guarantee his victory.

Input Format

The first line contains an integer n spells. the number of Merlin's

The second line contains n space-separated integers -Merlin's spell power levels.

The third line contains an integer m the number of Morgana's spells.

The fourth line contains m space-separated integers -Morgana's spell power levels.

Output Format

A single integer: the minimum mana required for Merlin to ensure victory in all possible cases.

Constraints

1 ≤ n, m ≤ 105

0 ≤ spell power ≤ 109

```python
def min_mana_to_win(n, merlin_spells, m, morgana_spells):
    max_morgana = max(morgana_spells)
    required_power = max_morgana + 1
    mana = 0

    for power in merlin_spells:
        if power < required_power:
            mana += (required_power - power)

    return mana

# Input reading
if __name__ == "__main__":
    n = int(input())
    merlin_spells = list(map(int, input().split()))
    m = int(input())
    morgana_spells = list(map(int, input().split()))
    
    print(min_mana_to_win(n, merlin_spells, m, morgana_spells))
```
