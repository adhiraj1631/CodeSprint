The crew of the starship Nebula Voyager must calibrate the warp drive by combining three mineral crystals. Each crysta emits a specific amount of energy, and the combined energ of three selected crystals must be as close as possible to the required target energy for a safe jump.

If multiple combinations are equally close to the target, choose the combination with the maximum total energy to avoid underpowering the drive.

Input Format

First line: An integer n - the number of crystals in the array

Second line: n space-separated integers values of each crystal the energy

Third line: An integer target - the required warp energy

Output Format

A single integer: the optimal total energy of a selected triplet (sum of three crystals) that is closest to the target.

If multiple sums are equally close, return the maximum such sum.

Example

Input 1:

4

-1023

3

Output 1:

4

Explanation:

Possible triplets and their sums:

[-1, 0, 2] → sum = 1-13-1| = 2

[-1, 0, 3] → sum = 2 - 13 - 2 = 1

[-1, 2, 3] → sum = 4-13-4 = 1

[0, 2, 3] → sum = 5-13-5 = 2

Closest sums to 3 are 2 and 4. Choose the maximum → 4.

Input 2:

4

10-257

13

Output 2:

13

Explanation:

The triplet [10, -2, 5] gives an exact match of 13. No closer sum exists.

8 Constraints

3 ≤ n ≤ 1000

-105 ≤ arr[i] ≤ 105

-107 ≤ target ≤ 107

```python
def closest_triplet_sum(n,arr,target):
  arr.sort()
  closest_sum=float('inf')
  min_diff=float('inf')
  for i in range(n-2):
    left,right=i+1,n-1
    while left<right:
      total=arr[i]+arr[left]+arr[right]
      diff=abs(target-total)
      if diff <min_diff or (diff==min_diff and total>closest_sum):
        closest_sum=total
        min_diff=diff
      if total<target:
        left+=1
      else:
        right-=1
  print(closest_sum)
n=int(input())
arr=list(map(int,input().split()))
target=int(input())
closest_triplet_sum(n,arr,target)
```
