Benny the pig has just taken a shower and is going to buy gifts for her family. Her friend Mike gives her a list of steps to follow, using the letters:

L (left)

R (right)

U (up)

D (down)

Every time Benny steps on a new spot, the snow turns to ice because she's wet.

If she steps on a spot she has already visited before, she slips and falls!

Your job is to count how many times Benny slips and falls as she follows Mike's instructions.

Input Format
A single string S consisting only of the characters L, R, U, and D.

(1≤5 ≤ 105)

Output Format

A single integer: the number of times Benny slips (i.e.,

revisits a position)

Movement Rules

Benny starts at position (0, 0).

Each movement changes her position by 1 unit in the given

direction.

Examples

Example 1

Input: LR

Output: 1

Explanation:

Start: (0, 0)

Move Left(-1.0)→ new spot (no slip)

←

Move Right → (0, 0) → already visited → slip!

Total slips = 1

Example 2

Input: UDUD

Output: 3

Explanation:

(0,0)→(0,1) (0,0)→(0,1) (0,0)

Visits same spots again and again.

Slips at:

(0,0) the first time revisited

(0,1) again

(0,0) again

Total slips = 3

Example 3

Input: RRRR

Output: 0

Explanation: All new spots → No slips.

```python
def count_slips(instructions):
    x, y = 0, 0
    visited = set()
    visited.add((x, y))
    slips = 0

    # Movement map
    moves = {'L': (-1, 0), 'R': (1, 0), 'U': (0, 1), 'D': (0, -1)}

    for step in instructions:
        dx, dy = moves[step]
        x += dx
        y += dy
        if (x, y) in visited:
            slips += 1
        else:
            visited.add((x, y))

    return slips

# Example usage
if __name__ == "__main__":
    S = input().strip()
    print(count_slips(S))
```
