Theme Park Ticket Queue Challenge

Problem Statement

You are managing the ticket queue for a popular theme park ride. Each visitor in line wants to buy a specific number of tickets. The queue operates as follows:

The visitor at the front of the line buys exactly 1 ticket, which takes 1 second.

If the visitor still needs more tickets, they go to the end of the line.

If the visitor is done buying tickets, they leave the queue.

Your task is to determine how many seconds it will take for the visitor who was initially at position k (0-indexed) to finish buying all their tickets.
Input Format

tickets: An array of integers where tickets[i] is the number of tickets the i-th person wants.

k: An integer representing the initial position of the target visitor in the queue.

Output Format

A single integer - the total time (in seconds) it takes for the visitor at index k to finish buying all their tickets.

Examples

Example 1:

Input:

tickets = [2, 3, 2]

k = 2

Output:

6

Explanation:
Time progression (each step = 1 second):

Queue: [2, 3, 2] → Visitor 0 buys 1 → [1, 3, 2]

Visitor 1 buys 1 - [1, 2, 2]

Visitor 2 buys 1 - [1, 2, 1]

Visitor 0 buys 1 → leaves → [2, 1]

Visitor 1 buys 1 - [1, 1]

Visitor 2 (now front) buys last ticket → done!

Total time = 6 seconds.

Example 2:

Input:

tickets = [5, 1, 1, 1] k = 0

Output: 8

Explanation:

Visitor 0 buys 1 - [4, 1, 1, 1] 2-4. Visitors 1, 2, 3 buy and leave 1415-8 Visitor 0.(alone.now).buys remaining 4
Example 2:

Input:

tickets [5, 1, 1, 1] k = 0

Output:

8

Explanation:

Visitor 0 buys 1-→ [4, 1, 1, 1] 2-4. Visitors 1, 2, 3 buy and leave → [4] 5-8. Visitor 0 (alone now) buys remaining 4 tickets.

Total time = 8 seconds.

Constraints

1 tickets.length ≤ 100

1 tickets[i] ≤ 100

0≤ k < tickets.length

```python
def time_to_buy_tickets(tickets, k):
    time = 0
    target = tickets[k]
    
    for i in range(len(tickets)):
        if i <= k:
            time += min(tickets[i], target)
        else:
            time += min(tickets[i], target - 1)
    
    return time
# Example 1
print(time_to_buy_tickets([2, 3, 2], 2))  # Output: 6

# Example 2
print(time_to_buy_tickets([5, 1, 1, 1], 0))  # Output: 8
```
