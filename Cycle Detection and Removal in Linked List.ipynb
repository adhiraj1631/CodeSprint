Cycle Detection and Removal in Linked List

Problem Statement

Preetesh is a software intern at a logistics company that tracks package deliveries using a singly linked list. Each node in the list represents a delivery record. However, due to occasional system glitches, the last delivery record may point back to a previous record, forming a loop in the tracking system.

This loop causes the system to enter an infinite cycle during delivery processing, disrupting operations.

Your task is to help Preetesh detect and remove the loop (if it exists). After fixing the issue, the linked list should remain in the same order, and the last node should correctly point to null.

Input Format

A singly linked list where each node contains an integer(delivery ID).

An integer pos:

If pos = 0, the last node points to null (no loop).

If pos > 0, the last node points to the node at the given 1-based index (creating a loop).

Output Format

Return true if.

There was no loop, or

A loop was successfully detected and removed, and the list is now linear and intact.

Return false if:

The loop could not be detected or removed properly.

Constraints

The linked list contains at most 105 nodes.

Node values are positive integers.

Use constant extra space (i.e., do not use hashing or auxiliary data structures). Example 1

Input

Linked list: 1 -> 3-> 4

pos = 2

Output: true

Explanation

The last node (4) points back to the second node (3),

creating a loop: 1-3-4-3-4-3-

After detecting and removing the loop, the list becomes: 1

→3→4→null

Example 2

Input

Linked list: 1->8-> 3 -> 4

pos = 0

Output. true

Explanation

The list is already linear and contains no loop.

```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
def detect_and_remove_loop(head):
  slow=fast=head
  while fast and fast.next:
    slow=slow.next
    fast=fast.next.next
    if slow==fast:
      break
  else:
      return True
  slow=head
  if slow==fast:
      while fast.next!=slow:
        fast=fast.next
  else:
      while slow.next!=fast.next:
        slow=slow.next
        fast=fast.next
  fast.next=None
  return True
def create_linked_list_with_loop(values,pos):
  if not values:
    return None
  head=Node(values[0])
  current=head
  loop_node=None
  for i in range(1,len(values)):
    current.next=Node(values[i])
    current=current.next
    if i==pos-1:
      loop_node=current
  if pos>0:
    current.next=loop_node
  return head
def print_linked_list(head):
  result=[]
  visited=set()
  while head:
    if head in visited:
      result.append("LOOP_STARTS_HERE")
      break
    visited.add(head)
    result.append(str(head.data))
    head=head.next
  print(" -> ".join(result))
values1=[1,3,4]
pos1=2
head1=create_linked_list_with_loop(values1,pos1)
print(detect_and_remove_loop(head1))

values2=[1,8,3,4]
pos2=0
head2=create_linked_list_with_loop(values2,pos2)
print(detect_and_remove_loop(head2))

    
    
```
