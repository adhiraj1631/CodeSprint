Problem Statement

Tamanna is a scientist monitoring a lab where temperatures fluctuate dangerously. She records temperatures in a stack as they occur. To prevent equipment damage, she needs to instantly know the lowest temperature observed so far.

However, due to faulty sensors, she sometimes has to discard the most recent reading.

Your task is to help Tamanna efficiently track the minimum temperature as she updates her stack of readings.

Input Format

The first line contains an integer n - the number of operations to perform.

The next n lines contain one of the following operations:

PUSH x: A new temperature reading x (in °C) is added.

POP: The latest temperature is discarded.(due to sensor error)
MIN: Output the current lowest temperature.

Output Format

For every MIN operation, output the lowest temperature in the stack.

I If the stack is empty when MIN is called, print EMPTY.

If a POP is called on an empty stack, ignore it.

Example 1: Sensor Malfunction

Input

6

PUSH 5

PUSH 2

MIN

DOD
MIN

POP

MIN

POP

Output

2

5

Explanation

Stack: [5, 2]→MIN is 2.

POP removes 2 Stack: [5] → MIN is 5.

Final POP empties the stack.

Example 2: Extreme Cold Wave

Input

7

PUSH-10

MIN

PUSH 5
MIN

POP

MIN

POP

Output

-10

-10

-10

Explanation

Stack: [-10]→MIN is -10.

Stack: [-10, 5] → MIN still.-10.

POP removes 5 Stack: [-10]→ MIN is -10.

Final POP clears the stack.

Constraints

1 ≤ n ≤ 105

Temperatures can be negative or positive integers.

```python
def process_operations(n, operations):
    stack = []
    min_stack = []
    results = []

    for op in operations:
        if op.startswith("PUSH"):
            _, val = op.split()
            val = int(val)
            stack.append(val)
            if not min_stack or val <= min_stack[-1]:
                min_stack.append(val)
        elif op == "POP":
            if stack:
                popped = stack.pop()
                if popped == min_stack[-1]:
                    min_stack.pop()
        elif op == "MIN":
            if min_stack:
                results.append(str(min_stack[-1]))
            else:
                results.append("EMPTY")
    
    return results

# Example usage
n = int(input())
operations = [input().strip() for _ in range(n)]
results = process_operations(n, operations)
print("\n".join(results))
```
