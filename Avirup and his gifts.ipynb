Problem Statement

After his birthday party, Avirup decides to build a tower of gifts in his room. Each time a friend arrives, they hand him a gift, and he stacks it on top of his growing tower. Some gifts are so special that he receives them more than once from different friends!

Throughout the day, Avirup sometimes removes the top gift to open it. At other times, he stands back and wonders: How many times does the topmost gift type appear in my entire tower right now?

Help Avirup to keep track of his gift tower and answer his curiosity whenever he asks.

Input Format

The first line contains an integer n - the number of operations.

The following n lines each contain one of the commands:
The following n lines each contain one of the commands:

PUSH X Add gift with label x (a string) to the top.

POP - Remove the topmost gift (if tower is not empty).

COUNT - Print the number of times the topmost gift appears in the tower. If the tower is empty, print "EMPTY"

Output Format

For each COUNT command, print:

The frequency of the topmost gift if the tower is not empty.

Otherwise, print "EMPTY".

Example 1

Input

7

PUSH Car

PUSH Teddy

PUSH Car

COUNT
POP

COUNT

POP

Output

2

1

Explanation

Stack: [Car, Teddy, Car] → Top is "Car" → Appears twice →

Output 2.

POP → Stack: [Car, Teddy] → Top is "Teddy" → Appears once

→Output 1.

Example 2

Input

5

PUSH Book

COUNT
POP

Output

1

EMPTY

Constraints

1 ≤ n ≤ 105

1≤ Length of gift label ≤ 20

Gift labels consist of lowercase/uppercase letters and digits

```python
def process_operations(n, operations):
    stack = []
    freq = {}
    results = []

    for op in operations:
        if op.startswith("PUSH"):
            _, x = op.split(maxsplit=1)
            stack.append(x)
            freq[x] = freq.get(x, 0) + 1
        elif op == "POP":
            if stack:
                top = stack.pop()
                freq[top] -= 1
                if freq[top] == 0:
                    del freq[top]  # Optional: remove from dict
        elif op == "COUNT":
            if not stack:
                results.append("EMPTY")
            else:
                top = stack[-1]
                results.append(str(freq[top]))
    
    return results

# Example usage
n = int(input())
operations = [input().strip() for _ in range(n)]
results = process_operations(n, operations)
print("\n".join(results))
```
