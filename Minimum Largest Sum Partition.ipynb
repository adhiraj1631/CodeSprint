You are in charge of packing a line of parcels into boxes. There are N parcels in a row, each with a specific weight.

Your task is to divide these parcels into at most K consecutive groups (each group representing a box). Your objective is to pack the parcels such that the heaviest box (i.e., the group with the largest total weight) is as light as possible.

Input Format

-The first line contains two integers: N - the number of parcels, K - the maximum number of boxes allowed

-The second line contains N space-separated integers: The weights of the parcels

Output Format

Print a single integer: the smallest possible weight of the heaviest box if the parcels are divided optimally.

Examples

Example 1

Input:

42

1234

Output:

6

Explanation:

We want to divide the parcels [1, 2, 3, 4] into 2 consecutive groups such that the maximum group sum is minimized:

Option 1: [1], [234] → Max = 9

Option 2: [12], [34] → Max = 7

Option 3: [123], [4] → Max = 6 - Optimal

Example 2

8

Explanation:

Possible divisions:

Option 1: [3], [18] → Max = 9

Option 2: [31], [8] → Max = 8 - Optimal

Constraints

1 ≤ N ≤ 105

1≤K≤N

1 ≤ weight ≤ 10

```python
def possible(weights, K, target):
    groups = 1
    current_sum = 0
    for weight in weights:
        if current_sum + weight <= target:
            current_sum += weight
        else:
            groups += 1
            current_sum = weight
            if groups > K:
                return False
    return True

# Input
N, K = map(int, input().split())
weights = list(map(int, input().split()))

# Binary search
low = max(weights)
high = sum(weights)
answer = high

while low <= high:
    mid = (low + high) // 2
    if possible(weights, K, mid):
        answer = mid
        high = mid - 1
    else:
        low = mid + 1

print(answer)
```
