In the annual Galactic Robot Racing Championship, each robot's speed is recorded as a unique score. The top three fastest robots receive special medals, while the rest are ranked numerically.

Your task is to determine each robot's rank based on their speed scores.

Input Format

An array score of length n, where score[i] represents the speed of the i-th robot.

All scores are unique.

Output Format

An array of strings where the i-th element is the rank of the i-th robot.

Special ranks:

1st place: "Gold Medal"

2nd place: "Silver Medal"

3rd place: "Bronze Medal
4th to nth: Their placement number as a string (e.g., "4", "5", etc.)

Examples

Example 1

Input

score = [50, 80, 30, 100]

Output

["Bronze Medal", "Silver Medal", "4", "Gold Medal"]

Explanation

Sorted by speed:

1001st "Gold Medal"

80-2nd "Silver Medal"

50→3rd "Bronze Medal"

304th "4"

Mapped back to original order:

score[0] = 50 → "Bronze Medal"

score[1] = 80 → "Silver Medal"
score[2] = 30 → "4" score[3] = 100 → "Gold Medal"

Example 2

Input

score = [23, 45, 67, 12, 89]

Output

["4", "Bronze Medal", "Silver Medal", "5", "Gold Medal"]

Explanation

Sorted by speed:

891st "Gold Medal"

672nd "Silver Medal"

45→3rd "Bronze Medal"

234th "4"

125th-"5"

Constraints

1 ≤ score.length ≤ 10*

0 ≤ score[i] ≤ 109

```python
def findRelativeRanks(score):
    n = len(score)
    
    # Step 1: Sort the scores with their original indices
    sorted_scores = sorted([(s, i) for i, s in enumerate(score)], reverse=True)
    
    # Step 2: Prepare result list
    result = [""] * n
    
    # Step 3: Assign ranks
    for rank, (s, i) in enumerate(sorted_scores):
        if rank == 0:
            result[i] = "Gold Medal"
        elif rank == 1:
            result[i] = "Silver Medal"
        elif rank == 2:
            result[i] = "Bronze Medal"
        else:
            result[i] = str(rank + 1)
    
    return result
score = [50, 80, 30, 100]
print(findRelativeRanks(score))
# Output: ["Bronze Medal", "Silver Medal", "4", "Gold Medal"]
```
