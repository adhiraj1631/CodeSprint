Problem Statement

In the mystical world of Dota2, two ancient factions - the Radiant and the Dire - are locked in a fierce battle for dominance.

The battlefield? A senate where their representatives take turns to:

Ban an opposing senator (removing their right to act in future rounds).

Or, if only their party remains, declare victory for their faction.

Senators take Jactions in the original order they appear and continue in rounds until one party has no remaining active members. Each senator will always act in the best interest of their party.

Your task: Given the initial order of senators, determine which party will win.

Input Format
Input Format

A single string senate of length n (1) ≤ n ≤ 10), consisting

only of characters:

'R' Radiant senator

'D' Dire senator

Output Format

A single string: either "Radiant" or "Dire" the victorious party. indicating

Example 1

Input

senate = "RDD"

Output

Dire

Explanation

Round 1:

R bans the first D → Senate becomes: [R, D]
R bans the first D → Senate becomes: [R, D]

Next D is banned → skipped

Remaining D bans R

Round 2:

Only one Dire senator remains and declares victory.

Example 2

Input

senate = "RRDDD"

Output

Radiant

Explanation

Round 1:

R bans a D

R bans the next D

D is skipped (banned)

D is skipped (banned)

D bans an R

Output

Radiant

Explanation

Round 1:

R bans a D

R bans the next D

D is skipped (banned)

D is skipped (banned)

D bans an R

Round 2:

R bans the last remaining D

Only Radiant senators remain → Radiant wins.

Constraints

1 ≤ senate.length ≤ 10*

senate contains only characters 'R' and 'D'

```python
from collections import deque

def predictPartyVictory(senate: str) -> str:
    n = len(senate)
    radiant = deque()
    dire = deque()

    # Store indices of each senator in their respective queues
    for i, s in enumerate(senate):
        if s == 'R':
            radiant.append(i)
        else:
            dire.append(i)

    # Simulate rounds
    while radiant and dire:
        r = radiant.popleft()
        d = dire.popleft()

        # The senator with the smaller index bans the other and gets re-added to the queue
        if r < d:
            radiant.append(r + n)  # Move to next round
        else:
            dire.append(d + n)

    return "Radiant" if radiant else "Dire"

print(predictPartyVictory("RDD"))      # Output: Dire
print(predictPartyVictory("RRDDD"))    # Output: Radiant
print(predictPartyVictory("DRRDRD"))   # Output: Dire
```
