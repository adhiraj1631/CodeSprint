It's Kim Jong Un's birthday, and his friend gave him a box of numbered cards. Each card has a number written on it. Kim Jong Un lines up the cards in a row to form an array.

But Kim Jong Un is a bit picky-he only likes the arrangement if the first card has a bigger number than the last card. He calls such an arrangement "beautiful."

Luckily, Kim Jong Un can swap any two cards as many times

as he wants, sone can arrange the cards in any order. Now, Kim Jong Un wonders: How many different ways can he arrange the cards so that the first card is greater than the last card?

But there's a catch! Kim Jong Un only cares about which numbers are at the first and last positions, not the order of the cards in between. So, two arrangements are considered the same if the same numbers are at the first and last positions.

Your task is to help Kim Jong Un count all such unique beautiful arrangements.

Input Format

The first line contains a single integer N - the number of cards.

The second line contains N space-separated integers - the numbers on the cards.

Output Format

Print a single integer - the number of different beautiful arrangements (i.e., number of distinct (first, last) pairs such that first > last).

Example 1

Input
Input

3

123

Output

3

Explanation

Unique numbers on the cards: 1, 2, 3

Valid beautiful pairs where first > last:

(2, 1)

(3, 1)

(3, 2)

Total beautiful arrangements = 3

Example 2

Input

4

5555

Output
Input

4

5555

Output

0

Explanation

All numbers are the same (5), so no first card can be greater than the last card.

No beautiful arrangements possible.

Constraints

2≤ N ≤ 105

1≤ card number ≤ 109

```python
from collections import Counter

def count_beautiful_arrangements(cards):
  freq=Counter(cards)
  unique_vals=sorted(freq.keys())
  total=0
  prefix_sum=0
  for i in range(len(unique_vals)-1,-1,-1):
    val=unique_vals[i]
    total+=freq[val]*prefix_sum
    prefix_sum+=freq[val]
  return total
N=int(input())
cards=list(map(int,input().split()))
print(count_beautiful_arrangements(cards))
```
