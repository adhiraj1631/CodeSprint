The Postfix-to-Prefix Transformer

Problem Statement

In the magical land of Algorithmia, the villagers rely on a powerful tool known as the Expression Transformer. This mystical device takes a valid postfix expression - where operators come after their operands (e.g., AB+) -and transforms it into a prefix expression - where operators come before their operands (e.g., +AB).

Your task is to describe and implement the transformation process using a stack - the secret behind the Expression Transformer's magic.

Input Format

A single string representing a valid postfix expression. The string contains only uppercase/lowercase letters (operands) and standard operators (+,-,*,/).

No spaces are included.

Output Format A single string: the corresponding prefix expression.

Rules for Transformation

Scan the postfix expression from left to right.

Use a stack to process the characters:

If the character is an operand, push it onto the stack.

If the character is an operator:

Pop the top two operands from the stack.

Combine them into a prefix expression using the operator

(e.g., op + left + right).

Push the resulting expression back onto the stack.

At the end, the stack will contain the final prefix expression.

Examples

Example 1

Input

AB+CD-*

Output *+AB-CD

Explanation:

AB+ →+AB

CD---CD

Then *+AB-CD → *+AB-CD

Example 2

Input

ABC/-AK/L-*

Output

*-A/BC-/AKL

Explanation:

ABC/--A/BC

AK/L---/AKL

Then -A/BC-/AKL→-A/BC-/AKL

Constraints

The postfix expression will always be valid. Length ≤ 10 characters.



```python
# Function to convert postfix to prefix
def postfix_to_prefix(postfix):
    stack = []

    # Iterate over each character in the postfix expression
    for char in postfix:
        if char.isalpha():  # Operand
            stack.append(char)
        else:  # Operator
            # Pop the top two operands
            operand2 = stack.pop()
            operand1 = stack.pop()
            # Form the prefix expression: operator + operand1 + operand2
            new_expr = char + operand1 + operand2
            # Push the new expression back to the stack
            stack.append(new_expr)

    # Final result is the only element in the stack
    return stack[-1]

# Input from user
postfix_expr = input()

# Convert and print the result
print(postfix_to_prefix(postfix_expr))

```
