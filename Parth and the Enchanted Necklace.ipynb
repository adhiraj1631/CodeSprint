Parth and the Enchanted Necklace

Problem Statement

In the magical land of Numeria, Parth has discovered an enchanted necklace composed of mystical beads. Each bead is linked to its neighbors in both directions, forming a doubly linked list. Every bead has a number engraved on it. However, some beads are cursed! Given the head of the doubly linked list and a cursed number (target), help Parth remove all beads with that cursed number to reveal the necklace's true power.

Your task is to remove all nodes from the doubly linked list that have the given target value and return the head of the modified list.

Input Format

head: The head node of a doubly linked list.

target: An integer value - the cursed number to be removed.
Each node contains:

val: Integer value of the bead.

prev: Pointer to the previous node.

next: Pointer to the next node.

Output Format

Return the head of the modified doubly linked list with all nodes containing the target value removed.

Examples

Example 1

Input:

head→122232124

target = 1

Output: head→234

Explanation: Beads with the value 1 are removed. Remaining list is: 2324

Example 2
Input:

head-232-12422

target = 2

Output:

head-32-124

Explanation: Beads with the value 2 are removed. Remaining

list is: 32-14

Example 3

Input:

head-555

target = 5

Output:

head (empty list)

Explanation: All beads are cursed and removed, leaving an empty list.

Constraints
The doubly linked list may contain 0 or more nodes.

Node values can be negative or positive integers. The list must maintain bidirectional links (prev and next pointers) after removal.

```python
class DoublyListNode:
  def __init__(self,val=0,prev=None,next=None):
    self.val=val
    self.prev=prev
    self.next=next
def remove_cursed_beads(head: DoublyListNode, target: int)-> DoublyListNode:
  current=head
  while current:
    next_node=current.next
    if current.val==target:
      if current.prev:
        current.prev.next=current.next
      if current.next:
        current.next.prev=current.prev
      if current==head:
        head=current.next
    current=next_node
  return head
def list_to_doubly_linked_list(values):
  if not values:
    return None
  head=DoublyListNode(values[0])
  prev=head
  for val in values[1:]:
    node=DoublyListNode(val)
    prev.next=node
    node.prev=prev
    prev=node
  return head
def doubly_linked_list_to_liost(head):
  result=[]
  while head:
    result.append(head.val)
    head=head.next
  return result
head=list_to_doubly_linked_list([1,2,2,2,3,2,1,2,4])
target=1
new_head=remove_cursed_beads(head,target)
print(doubly_linked_list_to_liost(new_head))

head=list_to_doubly_linked_list([2,3,2,1,2,4,2,2])
target=2
new_head=remove_cursed_beads(head,target)
print(doubly_linked_list_to_liost(new_head))

head=list_to_doubly_linked_list([5,5,5])
target=5
new_head=remove_cursed_beads(head,target)
print(doubly_linked_list_to_liost(new_head))
```
