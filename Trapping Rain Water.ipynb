After a heavy rainstorm, water gets trapped between the buildings in the city. The cityscape is represented by an array of non-negative integers, where each integer represents the height of a building. Each building has a uniform width of 1 unit.

At positions where the height is lower than the tallest buildings to its left and right, rainwater can be trapped.

Your task is to calculate the total amount of rainwater that can be trapped after the rain.

Input Format

The first line contains an integer n - the number of buildings.
