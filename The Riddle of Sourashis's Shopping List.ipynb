The Riddle of Sourashis's Shopping List

Problem Statement

Sourashis has a magical shopping list written as a linked list.

He wants to impress his friends by reversing any segment of the list using secret runes that mark the start and end positions of the segment.

Your task is to reverse a portion of the list from position left to right (inclusive). The list is singly linked and positions are 1-based indexed.

You must return the updated list after reversing the specified segment.

Input Format

head: A list (array) representing the linked list.

left: An integer representing the starting position (1-based index) of the segment to reverse.

right: An integer representing the ending position (1-based index) of the segment to reverse.
Output Format

result: A list (array) with the specified segment reversed.

Examples

Example 1

Input:

head = ["Bread", "Eggs", "Milk", "Cheese", "Tomatoes"]

left = 2

right = 4

Output:

["Bread", "Cheese", "Milk", "Eggs", "Tomatoes"]

Explanation:

Reverse the segment from position 2 to 4:

Original segment: "Eggs", "Milk", "Cheese" Reversed segment: "Cheese", "Milk", "Eggs" Resulting list: [*Bread", "Cheese", "Milk", "Eggs", "Tomatoes"]

Example 2

Input:
head = ["Bread", "Eggs", "Milk", "Cheese", "Tomatoes"] left = 1

right = 5

Output:

["Tomatoes", "Cheese", "Milk", "Eggs", "Bread"]

Explanation:

Reverse the entire list from position 1 to 5:

Original: "Bread", "Eggs", "Milk", "Cheese", "Tomatoes"

Reversed: "Tomatoes", "Cheese", "Milk", "Eggs", "Bread""

The list contains strings representing shopping items.

The list can have between 1 and 10 items.

```python
class ListNode:
    def __init__(self, val="", next=None):
        self.val = val
        self.next = next

# Convert list to linked list
def build_linked_list(items):
    dummy = ListNode()
    curr = dummy
    for item in items:
        curr.next = ListNode(item)
        curr = curr.next
    return dummy.next

# Convert linked list to list
def linked_list_to_list(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result

# Core function: reverse segment between left and right
def reverseBetween(head, left, right):
    if not head or left == right:
        return head

    dummy = ListNode("", head)
    prev = dummy

    for _ in range(left - 1):
        prev = prev.next

    # Reverse sublist
    curr = prev.next
    for _ in range(right - left):
        temp = curr.next
        curr.next = temp.next
        temp.next = prev.next
        prev.next = temp

    return dummy.next

# Example usage
head = ["Bread", "Eggs", "Milk", "Cheese", "Tomatoes"]
left = 2
right = 4

linked = build_linked_list(head)
result_head = reverseBetween(linked, left, right)
result_list = linked_list_to_list(result_head)
print(result_list)  # Output: ["Bread", "Cheese", "Milk", "Eggs", "Tomatoes"]
```
