Sort a Doubly Linked List - Deepshika's Challenge

Problem Statement

Deepshika, an enthusiastic coder at KIIT, loves solving tricky data structure problems. One day, she encounters a unique challenge: sorting a doubly linked list.

You're tasked with helping Deepshika sort a given doubly linked list in ascending order

A doubly linked list is a type of linked list where each node

contains three fields:

A value

A pointer to the previous node

A pointer to the next node

Input Format

The first line contains a single integer N nodes in the doubly linked list. the number of

The second line contains N space-separated integers - the values of the nodes in the order they appear in the list. Output Format

Print the sorted doubly linked list in ascending order. The node values should be separated using " <-> "to reflect the bidirectional links.

Examples

Example 1

Input:

5 3 8 14 20 17

Output:

3 <-> 8 <-> 14 <-> 17 <-> 20

Explanation:

The input list: 3 <-> 8 <-> 14 <-> 20 <-> 17 (biotonic)

Sorted output: 3 <-> 8 <-> 14 <-> 17 <-> 20

Example 2

Input:

4 9752

Output:

2 <-> 5 <-> 7 <-> 9

Explanation:

The list is strictly decreasing Sorted output: 2 <-> 5 <-> 7 <-> 9

Example 3

Input:

25 8 12 15 20

Output: 2 <-> 5 <-> 8 <-> 12 <-> 15 <-> 20 Explanation:

The list is already sorted in increasing order

Constraints 1≤ N ≤ 1000

Node values are integers (positive or negative) Explanation:

The list is strictly decreasing

Sorted output: 2 <-> 5 <-> 7 <-> 9

Wap in python with optimal time and space complexity

```python
class DoublyListNode:
    def __init__(self, val):

        self.val = val
        self.prev = None
        self.next = None

def build_doubly_linked_list(values):
    if not values:
        return None
    head = DoublyListNode(values[0])
    current = head
    for val in values[1:]:
        new_node = DoublyListNode(val)
        current.next = new_node
        new_node.prev = current
        current = new_node
    return head

def sort_doubly_linked_list(head):
    # Step 1: Convert linked list to array
    nodes = []
    current = head
    while current:
        nodes.append(current)
        current = current.next

    # Step 2: Sort nodes based on value
    nodes.sort(key=lambda node: node.val)

    # Step 3: Rebuild sorted doubly linked list
    for i in range(len(nodes)):
        nodes[i].prev = nodes[i - 1] if i > 0 else None
        nodes[i].next = nodes[i + 1] if i < len(nodes) - 1 else None

    return nodes[0]  # New head

def print_doubly_linked_list(head):
    result = []
    current = head
    while current:
        result.append(str(current.val))
        current = current.next
    print(" <-> ".join(result))

# Input Handling
N_and_values = input().strip().split()
if len(N_and_values) == 1:
    N = int(N_and_values[0])
    values = list(map(int, input().strip().split()))
else:
    N = int(N_and_values[0])
    values = list(map(int, N_and_values[1:]))

# Build, sort, and print
head = build_doubly_linked_list(values)
sorted_head = sort_doubly_linked_list(head)
print_doubly_linked_list(sorted_head)
```
