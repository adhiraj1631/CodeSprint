Problem Statement

In the ancient kingdom of Westeros, winter is coming! To keep every villager safe and warm, the king has ordered you, the royal engineer, to design a standard heater system.

Each heater has a fixed warm radius R. Your task is to determine the smallest possible radius R such that every house is within the warm range of at least one heater.

You are given two lists:

The positions of all houses on a straight road.

The positions of all heaters on the same road.

A house is considered "warmed" if the distance between it and at least one heater is less than or equal to R.

Input Format

houses: List of integers representing the positions of the houses.

heaters: List of integers representing the positions of the heaters.
Output Format

A single integer. the minimum radius required so that every house is within range of at least one heater.

Examples

Example 1

Input:

houses = [1, 2, 3]

heaters = [2]

Output:

1

Explanation:

The heater is at position 2.

A radius of 1 can warm houses at positions 1 (2-1), 2, and 3 (2+1).

All houses are within range.

Example 2

Input:
Input:

houses = [1, 5]

heaters = [2]

Output:

3

Explanation:

House at 1 → distance from heater at 2 is |1-2| = 1

House at 5 → distance from heater at 2 is 15-21 = 3

The required radius must be the maximum of these distances = 3

Constraints

1's houses.length, heaters.length < 10

1≤ houses[i], heaters[i] ≤ 109

Positions are all integers and may not be in sorted order

```python
def find_min_radius(houses, heaters):
    # Sort the positions of houses and heaters
    houses.sort()
    heaters.sort()

    radius = 0

    for house in houses:
        # For each house, find the distance to the nearest heater
        min_dist = float('inf')
        for heater in heaters:
            dist = abs(house - heater)
            min_dist = min(min_dist, dist)
        
        # The required radius is the maximum of all these minimal distances
        radius = max(radius, min_dist)

    return radius

# Example 1
houses1 = [1, 2, 3]
heaters1 = [2]
print(find_min_radius(houses1, heaters1))  # Output: 1

# Example 2
houses2 = [1, 5]
heaters2 = [2]
print(find_min_radius(houses2, heaters2))  # Output:
```
