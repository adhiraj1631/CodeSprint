Problem Statement

The cargo bay of the legendary airship Skyward Falcon is overflowing with crates, each labelled with a number. The ship's clever automaton wants to organize the crates so that the most common crate numbers are grouped together for quick access.

If two crate numbers appear the same number of times, the automaton prefers to arrange the lower-numbered crates first.

Your task is to help the automaton by sorting the list of crate numbers by their frequency in descending order. If two numbers have the same frequency, sort them in increasing numerical order.

Input Format

First line: Integer n- number of crates

Second line: n space-separated integers arr[]-crate labels

Output Format
Print the sorted crate labels in a single line, separated by spaces.

Examples

Example 1

Input

11

23245

12 2 3 3 3

12

Output

3333222

12

12 4 5

6

Explanation

Frequency map:

3-4 times

2-3 times

12-2 times

4,5-1 time each

Sorted order by frequency and value:

3(4x) 2 (3x) 12 (2x) 4 and 5. (1x in increasing.order
Example 2

Input

8

78768655

Output

55667788

Explanation

All elements appear twice. Since frequencies are equal, sort

numerically: 5, 6, 7, 8

Constraints

1 ≤ n ≤ 100,000

1 ≤ arr[i] ≤ 1,000,000,000

```python
from collections import Counter

def sort_crates_by_frequency(n,arr):
  freq=Counter(arr)
  sorted_items=sorted(freq.items(),key=lambda x: (-x[1],x[0]))
  result=[]
  for num,count in sorted_items:
    result.extend([num]*count)
  print(' '.join(map(str,result)))
n=int(input())
arr=list(map(int,input().split()))
sort_crates_by_frequency(n,arr)
```
