Problem Statement

RCB has discovered magical scroll containing a string made up only of parentheses: '(' and ')'. The scroll holds a secret message, but to reveal it, RCB must follow these rules:

Valid Parentheses String:

An empty string ""

A string in the form "(" + A + ")" where A is a valid parentheses string

Objective:

```python
def remove_outer_parentheses(s: str) -> str:
    result = []
    balance = 0
    start = 0

    for i, ch in enumerate(s):
        if ch == '(':
            if balance == 0:
                start = i  # mark the start of a new primitive
            balance += 1
        else:
            balance -= 1
            if balance == 0:
                # Remove the outermost parentheses
                result.append(s[start + 1:i])

    return ''.join(result)

# Example usage
s = input().strip()
print(remove_outer_parentheses(s))
```
