Monk and his friend Micro love playing with words. One day, Monk writes down a list of words, one after another.

For each word, he wants to know:

How many of the words that he wrote before are earlier in the dictionary than the current word?

Monk calls this number the "niceness value" of the word. Your task is to help Monk find the niceness value for every word in his listrand print them in order, one per line.

Input Format

The first line contains a single integer N- the number of words Monk wrote down.

The next N lines each contain a word (only lowercase English letters).

Output Format

Drint the niconoce.unlue.for.onch.sword in order.ono.nor.ling
Output Format

Print the niceness value for each word, in order, one per line.

Example

Input

5

bat

ant

cat

apple

dog

Output

0

0

2

1

4

Explanation
Explanation

For "bat": No words before it, so niceness is 0.

For "ant": Only "bat" is before it, but "bat" is not earlier in the dictionary than "ant", so niceness is 0.

For "cat": "bat" and "ant" are before it, and both are earlier in the dictionary, so niceness is 2.

For "apple": "bat", "ant", and "cat" are before it, but only "ant" is earlier in the dictionary, so niceness is 1.

For "dog": All previous words ("bat", "ant", "cat", "apple") are earlier in the dictionary, so niceness is 4.

Constraints

1≤ N ≤ 105

Each word contains only lowercase English letters (a-z)

Word length ≤ 100

```python
from sortedcontainers import SortedList

def compute_niceness(words):
  sorted_list=SortedList()
  results=[]
  for word in words:
    niceness=sorted_list.bisect_left(word)
    results.append(niceness)
    sorted_list.add(word)
  return results
n=int(input())
words=[input().strip() for _ in range(n)]
for niceness in compute_niceness(words):
  print(niceness)
```
