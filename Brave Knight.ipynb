Problem Statement

You are a brave knight navigating a treacherous dungeon filled with treasure rooms and trap rooms. Each room has a point value:

Positive for treasure

Negative for traps

-

Starting at the first room (index 0), you must reach the exit (last room) by jumping forward up to k rooms at each step.

Your goal is to maximize your total score - the sum of all rooms you land on.

Choose your path wisely, for some rooms will drain your strength!

Input Format

nums: A 0-indexed integer array where nums[i] represents the points in the i-th room.

k: An integer representing the maximum number of rooms you can jump forward at each step.
コ

U

Output Format

A single integer: the maximum score achievable when reaching the last room.

Example 1

Input

nums = [1, -2, -5, 7, -6, 4] k = 2

Output

10

Explanation

Start at index 0 (score = 1)

Jump to index 1 (-2) → total = -1

Jump to index 3 (7) → total = 6

Jump to index 5 (4) - total = 10

Path: 0-1-3-5

Score: 1 + (-2) + 7 + 4 = 10

Example 2

Input

← Back to Problems
Jump to index 3 (7) → total = 6

Jump to index 5 (4) → total = 10

Path: 0-1-3-5

Score: 1 + (-2) + 7 + 4 = 10

Example 2

Input

nums = [2, 3, -2, 5, -1, 4] k = 3

Output

14

Explanation

Best Path: 0-1-3-5

Score: 2+3+5+4=14

Constraints

1 ≤ nums.length ≤ 105

-10 ≤ nums[i] ≤ 104

1 ≤ k ≤ 104

```python
from collections import deque

def max_score(nums, k):
    n = len(nums)
    dp = [0] * n
    dp[0] = nums[0]

    # deque stores indices of the dp array, in decreasing order of dp values
    dq = deque()
    dq.append(0)

    for i in range(1, n):
        # Remove indices from the deque that are out of window [i-k, i-1]
        while dq and dq[0] < i - k:
            dq.popleft()

        # dp[i] = nums[i] + max(dp[j]) for j in [i-k, i-1]
        dp[i] = nums[i] + dp[dq[0]]

        # Maintain the deque in decreasing order of dp values
        while dq and dp[i] >= dp[dq[-1]]:
            dq.pop()

        dq.append(i)

    return dp[-1]
print(max_score([1, -2, -5, 7, -6, 4], 2))  # Output: 10
print(max_score([2, 3, -2, 5, -1, 4], 3))   # Output: 14
```
